```python
from glob import *

#assign profile0 file
files = glob(directory+'profile*0.data')

for i,fn in enumerate(files): 
    #read the different filename into profile
    profile = ascii.read(fn, header_start=4, data_start=5)
    
    #plot Hydrogen vs Mass
    plot(profile['mass'], profile['x'], linewidth=2, color=cm.jet( float(i)/len(files) ), label=i*10) 
    #plot(x,y,linewidth, color(each filename gets a different color), label the different filenames)

#axis and legend
xlabel(r'Mass $[M_\odot]$',fontsize=15)
ylabel(r'$^1H\; mass\; fraction$',fontsize=15)
legend(loc=4,labelspacing=0.1)
show()
```

```python
for i,fn in enumerate(files):  
    #read the different filename into profile
    profile = ascii.read(fn, header_start=4, data_start=5) 
    
    #plot Helium vs Mass
    plot(profile['mass'], profile['y'], '--', linewidth=2, color=cm.jet( float(i)/len(files) ), label='')
    
    #plot Hydrogen vs Mass
    plot(profile['mass'],profile['x'], linewidth=2, color=cm.jet( float(i)/len(files) ), label=i*10)

#axis and legend
xlabel(r'Mass $[M_\odot]$',fontsize=15)
ylabel(r'$^1H\; &\; ^4He\; mass\; fraction$',fontsize=15)
legend(loc=4,labelspacing=0.1)
show()
```

The plot shows the hydrogen (solid lines) and helium (dash lines) mass fraction vs solar mass. Looking at the plot, I can conclude that the star's core are powered by nuclear fusion in which hydrogen is converted into helium. The star's core reachs a hydrogen mass fraction of about .68 and a helium mass fraction of .32 after a certain mass excluding line 0. But as hydrogen is expelled, the core eventually becomes the core of a white dwarf with a full helium core having a helim mass fraction of 1 and a hydrogen mass fraction of 0.

# 3) Extra CRED (optional): The Hertzsprung Russel Diagram

PLOT the solar mass model and your new star on a diagram that has *effective temperature* for the x-axis and *log Luminosity* for the y-axis. This is called the Hertzsprung Russel Diagram and it is one of the most famous plots in all of astronomy. 

READ a little bit online about the Hertzsprung Russel or HR diagram and COMMENT in a markdown cell below your plot on the tracks your two models trace out in the HR diagram. What are we seeing? How are the similar or different? 

```python
#plot new star's Luminosity vs Temperature

#plot 1 solar mass
plot( history['log_Teff'], history['log_L'], "y" )

#plate 50 solar mass
plot( newHistory['log_Teff'], newHistory['log_L'], "r--" )

#axis and legend
xlabel('$\log T_{eff}(K)$',fontsize=15)
ylabel(r'$\log(L/L_\odot)$',fontsize=15)
legend(['1 solar mass','50 solar mass'], loc = 4)
show()
```

The plot shows the evolution track of a 1 solar mass star and a 50 solar mass star. 

For the 1 solar mass star, the hydrogen are fusing to form helium. Eventually, the hydrgoen fuel in the core runs out and fusion stops. The helium core contracts converting gravitational potential energy into thermal energy. The core becomes hot enough for the helium to fuse to form carbon. The outer layers begin to expand, cool, and the luminosity decreases. This expansion and cooling causes the effective temperature to drop. The energy from the shell-burning region will transport its energy using convection to the outer layers  eventually increasing the luminosity.

For the 50 solar mass star, the star consume its hydrogen core at a fast rate compare to smaller mass stars. Once the hydrogen fuel is used up, the core contracts due to gravity and heats up. This triggers helium-burning in the core. As a result, the effective temperature of the star drops as its outer layers expand. The increase in radius is balanced out by a decrease in temperature which makes the luminosity remain essentially constant.
